```python
auto=df[['Ship Mode','Segment','Postal Code','Region','Category','Sales','Quantity','Discount','Profit']]
auto
```

```python
shipmode = pd.get_dummies(auto['Ship Mode'], drop_first = False)
shipmode
```

```python
auto = pd.concat([auto, shipmode], axis = 1)
auto
```

```python
segment = pd.get_dummies(auto['Segment'], drop_first = False)
segment
```

```python
auto = pd.concat([auto, segment], axis = 1)
auto
```

```python
region=pd.get_dummies(auto['Region'],drop_first=False)
region
```

```python
auto=pd.concat([auto,region],axis=1)
auto
```

```python
category=pd.get_dummies(auto['Category'],drop_first=False)
category
```

```python
auto=pd.concat([auto,category],axis=1)
auto
```

```python
auto=auto.drop(['Ship Mode','Segment','Region','Postal Code','Category'],axis=1)
auto
```

```python
auto.to_csv("modified superstore data.csv")
```

```python
auto.corr()
```

```python
import seaborn as sns
plt.figure(figsize=(15,5))
sns.heatmap(auto.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(auto, train_size = 0.85, test_size = 0.15, random_state = 1)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
X_train =df_train[['Sales','Quantity','Discount','First Class','Same Day',
             'Second Class','Standard Class','Consumer','Corporate','Home Office','Central','East','South',
             'West','Furniture','Office Supplies','Technology']] # independent variable

Y_train = df_train['Profit'] # dependent variable
lr = LinearRegression()
model = lr.fit(X_train, Y_train)
```

```python
df_test['Profit']
```

```python
df_test
```

```python
data = df_test.iloc[0:1]
actual_profit = data.pop('Profit')
data
```

```python
predict_profit=model.predict(data)
print("Actual Profit",actual_profit)
print("predicted profit=",predict_profit)
```

```python
X_test =df_test[['Sales','Quantity','Discount','First Class','Same Day',
             'Second Class','Standard Class','Consumer','Corporate','Home Office','Central','East','South',
             'West','Furniture','Office Supplies','Technology']] # independent variable

Y_test = df_test['Profit']
```

```python
predicted_profit=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

# Assuming y_test is the actual labels and predicted_salary contains the predicted labels
r2 = r2_score(Y_test, predicted_profit)
accuracy_percentage = r2 * 100
print(f'Accuracy: {accuracy_percentage:.2f}%')
```
