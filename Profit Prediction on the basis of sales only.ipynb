```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Superstore.csv',usecols = ['Sales','Profit'],encoding='windows-1254')
dataset = dataset.dropna()
dataset
```

```python
X=dataset.iloc[:,0:1].values
Y=dataset.iloc[:,1:].values
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.15,random_state = 10)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
predicted_profit=model.predict(X_test)
predicted_profit
```

```python
predicted_profit = model.predict(X_test)
for i in range(len(X_test)):
  print("Sales",X_test.tolist()[i][0], "     Predicted Profit",predicted_profit[i][0], "     Actual Profit",Y_test.tolist()[i][0])
```

```python
plt.scatter(X_train, Y_train)
plt.show()
```

```python
plt.scatter(Y_test,model.predict(X_test))
plt.xlabel("Actual Profit")
plt.ylabel("Predicted profit")
plt.show()
```
